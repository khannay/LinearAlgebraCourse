# Least Squares to Fit Projectile Motion

```julia
using LinearAlgebra
using Plots
```

```julia
t=collect(0:0.2:10.0)
N=length(t)
h=1.0.+100.0*t.-9.8*t.^2;
hdata=h.+5.0*randn(length(h))
plot(t,h, lw=2.0, label="Actual Motion")
scatter!(t,hdata, label="")
title!("Projectile Motion")
ylabel!("Height (m)")
xlabel!("time (sec)")
```

```julia
A=[ones(N)'; t'; t'.^2]'
b=hdata;

A'A
```

```julia
sol=A'A\A'b
```

```julia
hfit=sol[1].+sol[2]*t.+sol[3]*t.^2;
plot(t,h, lw=2.0, label="H Real")
plot!(t,hfit, lw=2.0, label="H Fit")
scatter!(t,hdata, label="")
title!("Projectile Motion")
ylabel!("Height (m)")
xlabel!("time (sec)")
```
